```python
#  Разработка системы оценки вклада SNP в исследуемый признак (предполагается использование GWAS анализа и извлечение коэффициента бета1 из уравнения регрессии)

# Step 1: Define a function to calculate genotype effect
def calculate_genotype_effect(genotype):
    """
    This function calculates the genotype effect:
    - Homozygous reference (e.g., G/G): 0
    - Heterozygous (e.g., G/T): 0.5
    - Homozygous alternate (e.g., T/T): 1
    """
    if genotype.count('/') != 1:
        return None  # Return None if genotype format is invalid
    alleles = genotype.split('/')
    if alleles[0] == alleles[1]:
        return 1 if alleles[0] == "alt" else 0
    else:
        return 0.5
```

```python
# Step 2: Apply the genotype effect calculation to the dataset
merged_data["Genotype_Effect"] = merged_data["Генотип коровы"].apply(calculate_genotype_effect)

# Step 3: Calculate SNP contribution
merged_data["SNP_Contribution"] = merged_data["beta"] * merged_data["Genotype_Effect"]

# Step 4: Aggregate contributions for each cow and trait
# Group by cow ID and trait, then sum the contributions
aggregated_contributions = (
    merged_data.groupby(["ID_особи", "Признак"])["SNP_Contribution"]
    .sum()
    .reset_index()
    .rename(columns={"SNP_Contribution": "Total_Contribution"})
)

# Print the first few rows of the aggregated contributions for verification
aggregated_contributions.head()
```

```python

# Step 2: Filter relevant traits and remove NaN values
relevant_traits = ["Удой л/день", "Прирост веса кг/день"]
filtered_data = merged_genotypes[merged_genotypes["Признак"].isin(relevant_traits)]
filtered_data = filtered_data.dropna(subset=["Удой л/день", "Прирост веса кг/день"])

# Step 3: Separate Milk and Meat Contributions
milk_data = filtered_data[filtered_data["Признак"] == "Удой л/день"].copy()
meat_data = filtered_data[filtered_data["Признак"] == "Прирост веса кг/день"].copy()

# Ensure SNP details are included
required_columns = ["ID_особи", "chrom", "pos", "ref", "alt", "SNP_Contribution"]
milk_data = milk_data[required_columns].rename(columns={"SNP_Contribution": "Milk_Contribution"})
meat_data = meat_data[required_columns].rename(columns={"SNP_Contribution": "Meat_Contribution"})

# Step 4: Merge Milk and Meat Contributions
merged_snp = pd.merge(
    milk_data,
    meat_data,
    on=["ID_особи", "chrom", "pos", "ref", "alt"],
    how="inner"
)

# Step 5: Calculate Interaction Effect
merged_snp["Interaction_Effect"] = (
    merged_snp["Milk_Contribution"] * merged_snp["Meat_Contribution"]
)

# Step 6: Aggregate Interaction Effects at Animal Level
interaction_results = (
    merged_snp.groupby("ID_особи")["Interaction_Effect"]
    .sum()
    .reset_index()
    .rename(columns={"Interaction_Effect": "Total_Interaction_Effect"})
)

# Step 7: Rank Animals by Interaction Effect
interaction_results["Rank"] = interaction_results["Total_Interaction_Effect"].rank(ascending=False)

# Step 8: Sort Results
sorted_results = interaction_results.sort_values(
    by=["Total_Interaction_Effect"], ascending=False
).reset_index(drop=True)

# Display the top results
print(sorted_results.head(20))

# Save to CSV if needed
#sorted_results.to_csv("/path/to/interaction_results.csv", index=False)

```

```python
merged_genotypes = merged_genotypes.dropna(subset=["Удой л/день", "Прирост веса кг/день"]).reset_index(drop=True)

# Step 1: Filter relevant traits in aggregated_contributions
relevant_traits = ["Удой л/день", "Прирост веса кг/день"]
filtered_contributions = aggregated_contributions[aggregated_contributions["Признак"].isin(relevant_traits)]

# Step 2: Pivot aggregated contributions to separate milk and meat contributions
pivot_contributions = filtered_contributions.pivot(
    index="ID_особи",
    columns="Признак",
    values="Total_Contribution"
).reset_index()

# Rename columns for clarity
pivot_contributions.columns = ["ID_особи", "Milk_Contribution", "Meat_Contribution"]

# Step 3: Merge with production data to include actual production values
# Extract necessary columns from merged_data
production_data = merged_genotypes[["ID_особи", "Удой л/день", "Прирост веса кг/день"]].drop_duplicates()

# Merge contributions with production data
merged_results = pd.merge(
    pivot_contributions,
    production_data,
    on="ID_особи",
    how="left"
)

# Step 4: Handle NaN values and exclude negative contributions
# Replace NaNs with 0 for consistency
merged_results.fillna(0, inplace=True)

# Filter out rows with negative or zero contributions for milk and meat
merged_results = merged_results[
    (merged_results["Milk_Contribution"] > 0) & 
    (merged_results["Meat_Contribution"] > 0)
]

# Step 5: Calculate interaction effects
# Multiply contributions for milk and meat
merged_results["Interaction_Effect"] = (
    merged_results["Milk_Contribution"] * merged_results["Meat_Contribution"]
)

# Step 6: Rank animals by interaction effects
merged_results["Rank"] = merged_results["Interaction_Effect"].rank(ascending=False)

# Sort by Interaction Effect to prioritize genetic potential
sorted_results = merged_results.sort_values(
    by=["Interaction_Effect", "Milk_Contribution", "Meat_Contribution"],
    ascending=[False, False, False]
)

# Reset index for clean output
sorted_results.reset_index(drop=True, inplace=True)

# Display the top results
print(sorted_results.head())
```
